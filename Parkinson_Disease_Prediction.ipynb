
**Problem Statement:** Predict whether the person has parkinson disease or not.

Import the dependencies

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn import metrics
```

Import the data

```python
parkin_data = pd.read_csv(f'/content/Parkinsson disease.csv')
parkin_data.head()
```

Exploratory Data Analysis

```python
# Let us see the number of rows and columns
parkin_data.shape
```

```python
# Let us check whether the data is biased or not
parkin_data["status"].value_counts()
```

We can see that the data is baised, so when splittting the data for training and testing, we will do stratified sampling

```python
# Let us analyze the data
parkin_data.info()
```

We see all the values are numerical values

```python
# Let us check whether we have any missing values or not
parkin_data.isna()
```

```python
# Let us check whether we have missing values in the columns
parkin_data.isna().any()
```

```python
# Let us check whether we have any rows having the missing values
parkin_data.isna().any(axis = 1)
```

```python
# Let us see the statistical measure of the data
parkin_data.describe()
```

Now, we see that the data is not scaled, so we will use standard scale to scale the data

```python
# Let us see which data has high relation with the status
corr = parkin_data.corr()
corr.style.background_gradient(cmap='coolwarm')
```

We see that some has negative correlation and some has positive correlation with the status column.

**Split the data into predictor and target varaibes.**

```python
X = parkin_data.drop(columns = ['status','name'])
y = parkin_data['status']
print("The shape of X is ", X.shape)
print("The shape of y is ", y.shape)
```

Let us split the data into training and testing data-set

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size = 0.8, random_state = 2)
print("The shape of X_train is ", X_train.shape)
print("The shape of y_train is ", y_train.shape)
print("The shape of X_test is ", X_test.shape)
print("The shape of y_test is ", y_test.shape)
```

Data standardization

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
scaler.fit(X_train)
X_train = scaler.transform(X_train)
X_test = scaler.transform(X_test)
```

```python
print(X_train)
```

**Model Training**

```python
from sklearn import svm
clf = svm.SVC(kernel = 'linear')
clf.fit(X_train, y_train)
```

Model Evaluation

Accuracy score train data

```python
y_train_pred = clf.predict(X_train)


# Model Accuracy: how often is the classifier correct?
print("Accuracy:",metrics.accuracy_score(y_train, y_train_pred))
```

Accuracy score test data

```python
y_pred = clf.predict(X_test)

# Model Accuracy: how often is the classifier correct?
print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
```
